# 12 `set`s

- A `set` is a collection which is unordered and unindexed.
- In Python `set`s are written with curly brackets.
- **A `set` cannot have duplicated values**.

# 12.1 Create a `set`

```python
boys = {'Ali', 'Ahmad', 'Noor'}

print(boys)
```

```python
girls = {'Raya', 'Ola', 'Raya', 'Raya', 'Dana'}

print(girls)
```

```python
numbers = {1, 2, 3, 3, 2, 1}

print(numbers)
```

# 12.2 Access `set`s Data

- You cannot access items in a `set` by referring to an index, since sets are unordered the items has no index.
- You can use `for` loop to iterate over the `set`.

```python
for name in boys:
    print(name)
```

# 12.3 Is item `in` the `set`?

```python
print('Ali' in boys)
```

```python
print('Naser' in boys)
```

# 12.4 Modifying `set`s

```python
print(girls)
```

```python
girls.add('Hadeel')

print(girls)
```

# 12.5 Deleting Items

```python
girls.remove('Raya')

print(girls)
```

```python
girls.clear()
print(girls)
```

```python
del girls
print(girls)
```
